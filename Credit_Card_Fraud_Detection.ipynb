```python
# Dropping Unnamed column

CCF.drop("Unnamed: 0",axis=1,inplace=True)
CCF_test.drop("Unnamed: 0",axis=1,inplace=True)
CCF.head()
```

```python
# Converting date columns to datetime format

CCF['trans_date_trans_time']=pd.to_datetime(CCF['trans_date_trans_time'])
CCF['trans_date']=CCF['trans_date_trans_time'].dt.strftime('%Y-%m-%d')
CCF['trans_date']=pd.to_datetime(CCF['trans_date'])
CCF['dob']=pd.to_datetime(CCF['dob'])

CCF_test['trans_date_trans_time']=pd.to_datetime(CCF_test['trans_date_trans_time'])
CCF_test['trans_date']=CCF_test['trans_date_trans_time'].dt.strftime('%Y-%m-%d')
CCF_test['trans_date']=pd.to_datetime(CCF_test['trans_date'])
CCF_test['dob']=pd.to_datetime(CCF_test['dob'])

CCF.head()
```

```python
# Dropping fields not necessary for model building

drop_cols = ['cc_num','merchant','first','last','street','zip','trans_num','unix_time']

CCF1 = CCF.drop(drop_cols,axis=1)
CCF_test1 = CCF_test.drop(drop_cols,axis=1)

CCF1.info()
```

```python
# Checking Class Imbalance

classes_train=CCF1['is_fraud'].value_counts()
normal_share_train=classes_train[0]/CCF1['is_fraud'].count()*100
fraud_share_train=classes_train[1]/CCF1['is_fraud'].count()*100

classes_test=CCF_test1['is_fraud'].value_counts()
normal_share_test=classes_test[0]/CCF_test1['is_fraud'].count()*100
fraud_share_test=classes_test[1]/CCF_test1['is_fraud'].count()*100

plt.subplot(2,1,1)
plt.bar(['Non-Fraud','Fraud'], classes_train, color=['y','b'])
plt.title('Train')
plt.ylabel('Number of transactions')
plt.annotate("{0:.4}%".format(normal_share_train),(0.2, 0.5), xycoords='axes fraction')
plt.annotate("{0:.4}%".format(fraud_share_train),(0.7, 0.5), xycoords='axes fraction')

plt.subplot(2,1,2)
plt.bar(['Non-Fraud','Fraud'], classes_test, color=['y','b'])
plt.title('Test')
plt.ylabel('Number of transactions')
plt.annotate("{0:.4}%".format(normal_share_test),(0.2, 0.5), xycoords='axes fraction')
plt.annotate("{0:.4}%".format(fraud_share_test),(0.7, 0.5), xycoords='axes fraction')

plt.show()
```

```python
# Creating field for age on transaction date

CCF1['age_at_trans'] = CCF1['trans_date'] - CCF1['dob']
CCF1['age_at_trans'] = CCF1['age_at_trans'].astype('timedelta64[Y]')

CCF_test1['age_at_trans'] = CCF_test1['trans_date'] - CCF_test1['dob']
CCF_test1['age_at_trans'] = CCF_test1['age_at_trans'].astype('timedelta64[Y]')

CCF1.head()
```

```python
## Dropping variables not useful for model building

drop_cols = ['trans_date_trans_time','city','lat','long','job','dob','merch_lat','merch_long','trans_date']

CCF2 = CCF1.drop(drop_cols,axis=1)
CCF_test2 = CCF_test1.drop(drop_cols,axis=1)

CCF2.head()
```

```python
#Binary mapping on Gender

CCF2['gender'] = CCF2['gender'].map({'M': 1, 'F': 0})
CCF_test2['gender'] = CCF_test2['gender'].map({'M': 1, 'F': 0})

CCF2.head()
```

```python
CCF2[['amt','city_pop','age_at_trans','lat_dist','long_dist']].describe(percentiles=[.25, .5, .75, .90, .95, .99])
```

```python
# Creating dummy variables for Category

CCF3 = pd.get_dummies(CCF2,columns=['category'],drop_first=True)
CCF_test3 = pd.get_dummies(CCF_test2,columns=['category'],drop_first=True)

CCF3.info()
```

##Building Model

```python
# Dropping State

CCF4 = CCF3.drop('state',axis=1)
CCF_test4 = CCF_test3.drop('state',axis=1)
CCF4.info()
```

```python
# Splitting train and test dataset into X and y

X_train = CCF4.drop(['is_fraud'],axis=1)
y_train = CCF4['is_fraud']

X_test = CCF_test4.drop(['is_fraud'],axis=1)
y_test = CCF_test4['is_fraud']
```

Oversampling data to compensate for the high class imbalance in the data

```python
# Using ADASYN to oversample

from imblearn.over_sampling import ADASYN
X_train_res, y_train_res = ADASYN().fit_resample(X_train, y_train)

from collections import Counter
print(sorted(Counter(y_train_res).items()))
```

#Building a Random Forest Model

```python
# Importing required library

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(n_estimators=10, max_depth=5, max_features=12, min_samples_leaf=100, random_state=100, oob_score=True)
```

```python
rf.fit(X_train_res, y_train_res)
```

```python
rf.oob_score_
```

```python
log_regression = LogisticRegression()
log_regression.fit(X_train,y_train)
y_pred_proba = log_regression.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)

plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
imp_df = pd.DataFrame({
    "Varname": X_train.columns,
    "Imp": rf.feature_importances_
})
```

```python
imp_df.sort_values(by="Imp", ascending=False)
```

```python
y_train_pred = rf.predict(X_train_res)
y_train_pred[:10]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train_res, y_train_pred))
```

Predicting on Test set

```python
y_test_pred = rf.predict(X_test)
y_test_pred[:10]
```

```python
print(classification_report(y_test, y_test_pred))
```

Testing model on original data without oversampling

```python
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)

print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```
